#### Imports
- KMeans for the clustering algo
- matplotlib for reading and plotting images
- Counter to keep count of colors
- random to randomly get integers as number of colors later on

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from collections import Counter
import random
```

#### Function to return the hex value of a color given

```python
def RGB2HEX(color):
    return "#{:02x}{:02x}{:02x}".format(int(color[0]), int(color[1]), int(color[2]))
```

#### Color Identification is done using KMeans Algorithm
- And KMeans expects flattened array as input during its fit method. 
- Thus, we reshaped the image by multiplying its two dimensions.
- Then, we applied KMeans to first fit and then predict on the image to get the results. 
- Then, the cluster colors are identified an arranged in the correct order. 
- We plot the colors as a pie chart.


#### Function to read and return image given its path

```python
def get_image(image_path):
    return plt.imread(image_path)
```

### The most important method which does all the works
- Modifies Image to reshape it and feeds to KMeans
- Get the RGB Values of colors and prints it by the inbuilt method (fit_predict)
- Order colors by iterating through keys and saving hex colors then rgb colors
- To show the pie chart we again use matplotlib to show the number of colors as given to this function

```python
def get_colors(image_path, number_of_colors, show_chart):
    fig, (ax1, ax2) = plt.subplots(1, 2)
    image = get_image(image_path)
    ax1.imshow(image)
    
    modified_image = image.reshape(image.shape[0]*image.shape[1], 3)
    clf = KMeans(n_clusters = number_of_colors)
    labels = clf.fit_predict(modified_image)
    #print(labels)
    
    counts = Counter(labels)
    center_colors = clf.cluster_centers_
    ordered_colors = [center_colors[i] for i in counts.keys()]
    hex_colors = [RGB2HEX(ordered_colors[i]) for i in counts.keys()]
    rgb_colors = [ordered_colors[i] for i in counts.keys()]

    if (show_chart):
        ax2.pie(counts.values(), labels = hex_colors, colors = hex_colors)
    
    return rgb_colors
```

#### Calling the function and getting our desired Output

```python
for i in range(1, 13):
    rgb_colors = get_colors(image_path=f'{i}.jpeg', number_of_colors=random.randint(3, 11), show_chart=True)
    
```
